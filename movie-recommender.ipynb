**dictionary to string**

```python
import ast
```

```python
def convert(dict):
    l=[]
    for i in ast.literal_eval(dict):
        l.append(i['name'])
    return l
```

```python
data['genres']=data['genres'].apply(convert)
```

```python
data['keywords']=data['keywords'].apply(convert)
```

**Getting only top three actors from cast**

```python
def convert2(dict):
    l=[]
    count=0
    for i in ast.literal_eval(dict):
        if(count<3):
            l.append(i['name'])
            count+=1
        else:
            break
    return l
```

```python
data['cast']=data['cast'].apply(convert2)
```

```python
data.head(3)
```

**Getting only directors from crew**

```python
def convert3(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            l.append(i['name'])
            break
    return l
```

```python
data['crew']=data['crew'].apply(convert3)
```

```python
data.head(3)
```

```python
data['overview']=data['overview'].apply(lambda x:x.split())
```

```python
data.head(3)
```

```python
data['genres']=data['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data['keywords']=data['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data['overview']=data['overview'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data['cast']=data['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data['crew']=data['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data['tags']=data['genres']+data['keywords']+data['overview']+data['cast']+data['crew']
```

```python
data.drop(['genres','keywords','overview','cast','crew'],axis=1,inplace=True)
```

```python
data['tags']=data['tags'].astype('str')
```

```python
data.head(3)
```

```python
data['tags']=data['tags'].apply(lambda x:x.lower())
```

```python
data.head(3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(data['tags']).toarray()
```

```python
vector.shape
```

```python
print(cv.get_feature_names_out())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
data['tags']=data['tags'].apply(stem)
```

```python
data.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
simillar=cosine_similarity(vector)
```

```python
simillar.shape
```

```python
sorted(list(enumerate(simillar[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=data[data['title']==movie].index[0]
    distances=simillar[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
       print(data.iloc[i[0]]['title'])
    
```

```python
recommend('Avatar')
```
